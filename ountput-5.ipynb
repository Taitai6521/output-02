```python
x = np.linspace(-10, 10, 10)
fig,axes = plt.subplots(figsize=(10,10))

axes.plot(x, x**2)
plt.savefig('quadlatic function')
plt.xlabel('x_axis')
plt.ylabel('y_axis')
plt.title('quadlatic function')
```

```python
fig, axes = plt.subplots(2,1,figsize=(10,10))
x = np.linspace(-10,10,100)
axes[0].plot(x, np.cos(x),label='cosine')
axes[1].plot(x, np.sin(x),label='sine')
axes[0].set_title('cos(x)')
axes[1].set_title('sin(x)')
axes[0].set_xlabel('x')
axes[0].set_ylabel('y')
axes[1].set_xlabel('x')
axes[1].set_ylabel('y')
fig.legend()
fig.tight_layout()
fig.savefig('cosine,sine')
```

```python
from matplotlib.backends.backend_pdf import PdfPages
pdf = PdfPages('pdf_files')
fig, axes = plt.subplots()
x= np.linspace(-10,10)
axes.plot(x,np.log10(x))
pdf.savefig(fig)
pdf.close()


```
